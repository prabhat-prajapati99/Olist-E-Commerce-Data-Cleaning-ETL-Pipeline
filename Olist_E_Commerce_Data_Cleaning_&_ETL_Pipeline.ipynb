> **Geolocation Table: Mean Spatial Aggregation**



---



```python
# 1. Group by zip code and get the average lat/lng
# We also take the 'first' city and state name found for that zip code
Geolocation_cleaned = Geolocation.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean',
    'geolocation_city': 'first',
    'geolocation_state': 'first'
}).reset_index()

# 2. Check the new size
print(f"Original rows: {len(Geolocation)}")
print(f"Cleaned rows: {len(Geolocation_cleaned)}")
print(f"Duplicates remaining: {Geolocation_cleaned.duplicated().sum()}")
```



---



> **String Normalization: Text Cleaning & Accents Removal Function**



> **Entity Synchronization: Matching Customer and Geolocation Cities**




---




```python
# Remove accents and special characters from city names
# This ensures "são paulo" becomes "sao paulo"
Geolocation_cleaned['geolocation_city'] = (
    Geolocation_cleaned['geolocation_city']
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
    .str.lower()
    .str.strip()
)

# Do the same for the customer file to be 100% sure they match
customer_cleaned['customer_city'] = (
    customer_cleaned['customer_city']
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
    .str.lower()
    .str.strip()
)
```

```python
Geolocation_cleaned.to_csv('Geolocation_cleaned.csv', index=False)
```

> **Reviews Table: Metadata Sanitization & Null Comment Placeholders**



---




```python
# Fill missing text with a standard placeholder
Reviews['review_comment_title'] = Reviews['review_comment_title'].fillna('no_title')
Reviews['review_comment_message'] = Reviews['review_comment_message'].fillna('no_comment')
```

```python
Reviews.head()
```

```python
import unicodedata  # <--- This fixes the NameError

def clean_special_characters(series):
    # 1. Convert to string and replace hidden line breaks (\r, \n) with spaces
    series = series.astype(str).str.replace(r'[\r\n]+', ' ', regex=True)

    # 2. Normalize Portuguese accents
    def remove_accents(text):
        normalized = unicodedata.normalize('NFKD', text)
        return "".join([c for c in normalized if not unicodedata.category(c).startswith('M')])

    series = series.apply(remove_accents)

    # 3. Keep only letters, numbers, and spaces
    series = series.str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

    # 4. Standardize spaces and make it lowercase
    series = series.str.strip().str.lower()

    return series

# Make sure 'reviews' uses lowercase 'r' to match your original variable name
Reviews['review_comment_message'] = clean_special_characters(Reviews['review_comment_message'])
```
